```python
import pandas as pd

#loading data from Wikipedia
wikipage = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

#saving the table as dataframe
toronto = wikipage[0]

#printing the first 5 rows of the dataframe
toronto.head()
```

```python
#removing rows without Borough data
toronto = toronto[toronto['Borough'] != 'Not assigned']

#printing the first 5 rows of the dataframe
toronto.head()
```

```python
#checking for repeated postal codes
if len(toronto['Postal Code']) == len(toronto['Postal Code'].unique()):
    print("There's no repeated Postal Code on the list")
else:
    repeated = len(toronto['Postal Code']) - len(toronto['Postal Code'].unique())
    print("There's " + repeated + "Postal Codes")    
```

```python
import geocoder # import geocoder

toronto['Latitude'] = ""
toronto['Longitude'] = ""

for index, row in toronto.iterrows():
    postalcode = (row['Postal Code'])
    print("Checking", ('{}, Toronto, Ontario'.format(postalcode)))

    # initialize variable to None
    lat_lng_coords = None

    # loop until get the coordinates
    while(lat_lng_coords is None):
        coord = geocoder.arcgis('{}, Toronto, Ontario'.format(postalcode))
        lat_lng_coords = coord.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    
    print(latitude, longitude)

    toronto.loc[index,['Latitude']] = latitude
    toronto.loc[index,['Longitude']] = longitude

    print("Row ", index, "is done")
```
